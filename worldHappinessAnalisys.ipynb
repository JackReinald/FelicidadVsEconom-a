```python
import matplotlib.pyplot as plt

# Averiguar la categoría que más afecta a Score mediante matriz de confusión
correlationMatrix = dfLimpio.select_dtypes(include="number").corr() # Selecciona solo las columnas numéricas y saca correlación

plt.figure(figsize=(12, 8))
sns.heatmap(correlationMatrix, annot=True, cmap="seismic", fmt=".2f")
plt.title("Matriz de correlacion")
plt.show()
```

```python
# Misma matriz de correlación, ahora con plotly
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Obtener las etiquetas para el eje x e y
labels = correlationMatrix.columns.tolist()

# Crear la figura con Plotly
fig = go.Figure(data=go.Heatmap(
    z=correlationMatrix.values,
    x=labels,
    y=labels,
    colorscale='RdBu',
    zmid=0, # Centrar el colorscale en 0 para la correlación
    colorbar=dict(title='Correlación')
))

fig.update_layout(
    title='Matriz de Correlación',
    xaxis_side='bottom',
    xaxis=dict(
        side='top',
        tickangle=-45, # Ángulo para las etiquetas del eje x (diagonal)
        tickfont=dict(size=13) # Opcional: ajustar el tamaño de la fuente
    ),
    height=800,
    width=900,
    annotations=[ go.layout.Annotation(
        x=col,
        y=row,
        text=f'{correlationMatrix.iloc[row, col]:.2f}',
        showarrow=False,
        font=dict(size=11.5),
        # bgcolor='rgba(255, 255, 255, 0.8)' if -0.2 < correlationMatrix.iloc[row, col] < 0.2 else 'rgba(0, 0, 0, 0.8)', # Fondo blanco para valores cercanos a 0, negro para el resto
        font_color='black' if -0.3 < correlationMatrix.iloc[row, col] < 0.3 else 'white' # Texto negro para valores cercanos a 0, blanco para el resto
    ) for row in range(correlationMatrix.shape[0]) for col in range(correlationMatrix.shape[1]) ]
)

# Mostrar la figura
fig.show()
```

```python
import numpy as np
# Filtrar solo aquellas correlaciones que sean mayores a +- 0.8
threshold = 0.8
mask = ( correlationMatrix.abs() > threshold) & (correlationMatrix != 1)
highCorrelations = correlationMatrix[mask].stack().reset_index()
highCorrelations

```

```python
# Se limpian los duplicados
highCorrelations = highCorrelations[highCorrelations["level_0"] < highCorrelations["level_1"]]
highCorrelations
```

```python
# Obtengo los nombres de las filas de highCorrelations (que son las categorías del dataframe original)
rowNames= highCorrelations.unstack().unique()
rowNames = rowNames[:6]
print(rowNames)
```

```python
# Como necesito saber si los datos siguen una distribución normal para averiguar el tipo de normalización que se necesita,
# hace falta realizar un histograma para las columnas correlacionadas y ver si este toma forma de campana (o no).

# Calculo la cantidad de barras que componen el histograma
bins = int(1+ np.log2(len(dfLimpio)))
bins
```

```python
# Se muestran los histogramas
for row in rowNames: 
    
    plt.figure()
    plt.hist(dfLimpio[row], bins)
    plt.title(row)

    
plt.show()
```

```python
nationData = dfLimpio
for row in highCorrelations.itertuples(index=False):
    
    lvl0, lvl1 = row[0], row[1]
    plt.figure()
    plt.scatter(nationData[lvl0], nationData[lvl1])
    plt.xlabel(lvl0)
    plt.ylabel(lvl1)
    plt.tight_layout()
    titulo = lvl0 + " VS " + lvl1
    plt.title(titulo)  
    
plt.show()
```

```python
from sklearn.preprocessing import RobustScaler


# copia= dfLimpio

# copia["Perceptions of corruption"] = dfLimpio["Perceptions of corruption"] * 100
# copia

# Escalado robusto del dataset para apropiada comparación en los gráficos de dispersión:
scaler = RobustScaler()
dfLimpioSoloNumeros = dfLimpio.select_dtypes(include="number")    # Exclusión de columnas con registros que no sean numéricos
dfPrepared = scaler.fit_transform(dfLimpioSoloNumeros)

# Se convierte nuevamente a dataframe
dfPrepared = pd.DataFrame(dfPrepared, columns=dfLimpioSoloNumeros.columns)


```

```python
# Comparación entre gráficas
index=1
for row in highCorrelations.itertuples(index=False):
    
    lvl0, lvl1 = row[0], row[1]
    plt.figure(figsize=(30, 7), num=index)

    plt.subplot(1,2, 1)
    plt.scatter(dfLimpio[lvl0], dfLimpio[lvl1], label="Original", color="blue")
    plt.xlabel(lvl0)
    plt.ylabel(lvl1)
    plt.legend()
    plt.grid(True)

    
    plt.subplot(1,2, 2)
    plt.scatter(dfPrepared[lvl0], dfPrepared[lvl1], label="Escaled", color="orange")
    
    plt.xlabel(lvl0)
    plt.ylabel(lvl1)
    plt.legend()
    plt.grid(True)

    titulo = lvl0 + " VS " + lvl1
    plt.title(titulo)  

    index += 1
plt.show()
```

```python
# Vector con los titulos con los análisis a hacer
titulos = ["Healthy life expectancy at birth", "Freedom to make life choices", "Perceptions of corruption", "Food Consumer Price Inflation", "GDP deflator Index growth rate"]

agrupacionPorContinente = dfLimpio.groupby("Continent")

for titulo in titulos:
    analisisPorTitulo = agrupacionPorContinente[titulo].mean().sort_values(ascending=False).reset_index()

    if( (analisisPorTitulo[titulo].mean() >= 0) & (analisisPorTitulo[titulo].mean() <= 1)):
        analisisPorTitulo[titulo] = analisisPorTitulo[titulo] * 100

    plt.figure(figsize=(11,7))
    plt.bar(x=analisisPorTitulo["Continent"], height=analisisPorTitulo[titulo])
    plt.ylabel("Percentage (%)")
    plt.title(titulo)
    plt.show()



```
